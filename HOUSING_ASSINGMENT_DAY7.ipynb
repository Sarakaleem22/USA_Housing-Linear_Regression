```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house = pd.read_csv("USA_Housing.csv")
```

```python
house.head()
```

```python
house.info()
```

```python
#no missing data
#Here the Label is Price and since it is numeric - apply Regression and since there are multiple feature - apply multivariate regression
house.corr()
```

```python
features = house.iloc[:,0:5].values
label = house.iloc[:,[5]].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

for i in range(1,5001):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size = 0.2,
                                                random_state=i)
    model = LinearRegression()
    model.fit(X_train,y_train)
    training_score = model.score(X_train,y_train)
    testing_score = model.score(X_test,y_test)
    
    
    if testing_score > training_score:
        print("Training Score {} Testing Score {} for Random State {}".format(training_score,testing_score,i))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size = 0.2,
                                                random_state=6)
```

```python
finalModel = LinearRegression()
finalModel.fit(X_train,y_train)
```

```python
print(finalModel.score(X_train,y_train))
print(finalModel.score(X_test,y_test))
```

```python
AvgAreaIn = float(input("Enter Avg Area Income: "))
AvgAreaHouseAge = float(input("Enter Avg Area House Age:"))
AvgAreaNoRooms  = float(input("Enter Avg area no of Rooms: "))
AvgAreaNoBedrooms = float(input("Enter the Avg Area no of bedrooms: "))
AreaPopulation = float(input("Enter the Area Population: "))
featureInput = np.array([[AvgAreaIn,AvgAreaHouseAge,AvgAreaNoRooms,AvgAreaNoBedrooms,AreaPopulation]])
price = finalModel.predict(featureInput)
print("Predicted Price is ",price)
```
